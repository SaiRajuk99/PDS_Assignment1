```python
import pandas as pd

data = pd.read_csv("raw_studentsdata.csv")
```

```python
clean_data = data.replace({'test preparation course': {'completed':'1','none':'0'}})
clean_data.head(10)
```

```python
clean_data.to_csv('clean_studentsdata.csv')
```

```python
df = pd.DataFrame(clean_data)
```

#Bar representation of Scores Genderwise

```python
import seaborn as sb
import matplotlib.pyplot as mpp

mean_testscores = df[['math score', 'reading score', 'writing score', 'gender']].groupby('gender').mean()
mean_testscores.plot(kind='bar', rot=0)
mpp.title('Mean Test Scores Gender wise')
mpp.xlabel('Gender')
mpp.ylabel('Mean Test Score')
mpp.show()
```

```python
sb.scatterplot(x='reading score',y='math score',hue='gender',data=df)
mpp.title('Reading Score Vs Math Score')
mpp.xlabel('Reading Score')
mpp.ylabel('Math Score')
```

```python
mpp.figure(figsize=(15,5))
mpp.subplot(1,3,2)
sb.kdeplot(x='math score',hue='gender',data=df,fill=True)
mpp.title('Density Curve of Math Score')
mpp.xlabel('Math Score')
mpp.ylabel('Density')

mpp.figure(figsize=(15,5))
mpp.subplot(1,3,2)
sb.kdeplot(x='reading score',hue='gender',data=df,fill=True)
mpp.title('Density Curve of reading Score')
mpp.xlabel('Reading Score')
mpp.ylabel('Density')

mpp.figure(figsize=(15,5))
mpp.subplot(1,3,2)
sb.kdeplot(x='writing score',hue='gender',data=df, fill=True)
mpp.title('Density Curve of writing Score')
mpp.xlabel('Writing Score')
mpp.ylabel('Density')
```

```python
sb.scatterplot(x='reading score',y='test preparation course',hue='gender',data=df,)
mpp.title('Test Preparation course Vs Reading score')
mpp.xlabel('Reading score')
mpp.ylabel('Test Preparation course')
mpp.show()
```

```python
import matplotlib.pyplot as mpp
mpp.figure(figsize=(10,5))
mpp.subplot(1,3,1)
mpp.hist(df['reading score'],bins=10, color="skyblue",edgecolor='black')
mpp.title('Reading Score vs Count')
mpp.xlabel('Reading score')
mpp.ylabel('Count')

mpp.figure(figsize=(10,5))
mpp.subplot(1,3,2)
mpp.hist(df['math score'],bins=10,color="lightgreen",edgecolor='black')
mpp.title('Math Score vs Count')
mpp.xlabel('Math score')
mpp.ylabel('Count')

mpp.figure(figsize=(10,5))
mpp.subplot(1,3,3)
mpp.hist(df['writing score'],bins=10,color="yellow",edgecolor='black')
mpp.title('Writing Score vs Count')
mpp.xlabel('Writing score')
mpp.ylabel('Count')
```

```python
mpp.figure(figsize=(8, 6))
df['gender'].value_counts().plot(kind='pie', autopct='%1.1f%%' )
mpp.title("Pie Chart: Gender Balance")
```

```python
df['test preparation course']=df['test preparation course'].astype(int)
sb.heatmap(df[['test preparation course','math score','reading score','writing score']].corr(), annot=True, cmap='coolwarm')
mpp.xticks(rotation=45)
mpp.yticks(rotation=0)
```
