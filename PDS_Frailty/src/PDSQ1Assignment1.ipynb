```python
import pandas as pd

data = pd.read_csv("raw_frailtydata.csv")
```

```python
clean_data = data.replace({'Frailty': {'Y':'1','N':'0'}})
clean_data.head(10)
```

```python
clean_data.to_csv('clean_frailtydata.csv')
```

```python
df = pd.DataFrame(clean_data)
df
```

```python
df.describe()
```

```python
df.corr()
```

```python

mgrip_frailty_y = df[df['Frailty'] == 1]['Grip strength'].mean()
mgrip_frailty_n = df[df['Frailty'] == 0]['Grip strength'].mean()

sd_grip_frailty_y = df[df['Frailty'] == 1]['Grip strength'].std()
sd_grip_frailty_n = df[df['Frailty'] == 0]['Grip strength'].std()

print("Means:")
print("Frailty = 1:", mgrip_frailty_y)
print("Frailty = 0:", mgrip_frailty_n)

print("\nStandard Deviations:")
print("Frailty = 1:", sd_grip_frailty_y)
print("Frailty = 0:", sd_grip_frailty_n)


```

```python

mheight_frailty_y = df[df['Frailty'] == 1]['Height'].mean()
mheight_frailty_n = df[df['Frailty'] == 0]['Height'].mean()
mweight_frailty_y = df[df['Frailty'] == 1]['Weight'].mean()
mweight_frailty_n = df[df['Frailty'] == 0]['Weight'].mean()
mage_frailty_y = df[df['Frailty'] == 1]['Age'].mean()
mage_frailty_n = df[df['Frailty'] == 0]['Age'].mean()


sd_height_frailty_y = df[df['Frailty'] == 1]['Height'].std()
sd_height_frailty_n = df[df['Frailty'] == 0]['Height'].std()
sd_weight_frailty_y = df[df['Frailty'] == 1]['Weight'].std()
sd_weight_frailty_n = df[df['Frailty'] == 0]['Weight'].std()
sd_age_frailty_y = df[df['Frailty'] == 1]['Age'].std()
sd_age_frailty_n = df[df['Frailty'] == 0]['Age'].std()

print("Means:")
print("Height - Frailty = Y:", mheight_frailty_y)
print("Height - Frailty = N:", mheight_frailty_n)
print("Weight - Frailty = Y:", mweight_frailty_y)
print("Weight - Frailty = N:", mweight_frailty_n)
print("Age - Frailty = Y:", mage_frailty_y)
print("Age - Frailty = N:", mage_frailty_n)

print("\nStandard Deviations:")
print("Height - Frailty = Y:", sd_height_frailty_y)
print("Height - Frailty = N:", sd_height_frailty_n)
print("Weight - Frailty = Y:", sd_weight_frailty_y)
print("Weight - Frailty = N:", sd_weight_frailty_n)
print("Age - Frailty = Y:", sd_age_frailty_y)
print("Age - Frailty = N:", sd_age_frailty_n)
```

```python
import numpy as nmp
from scipy import stats

grip_strength = df['Grip strength'].values
frailty = df['Frailty'].values

grip_strength_frailty_0 = grip_strength[frailty == 0]
grip_strength_frailty_1 = grip_strength[frailty == 1]

t_statistic, p_value = stats.ttest_ind(grip_strength_frailty_0, grip_strength_frailty_1)

print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

```python


print("Descriptive Statistics:")
print("Height - Mean:-", nmp.mean(df['Height']), "  Standard Deviation:", nmp.std(df['Height']))
print("Weight- Mean:-", nmp.mean(df['Weight']), "  Standard Deviation:", nmp.std(df['Weight']))
print("Age -Mean:-", nmp.mean(df['Age']), "  Standard Deviation:", nmp.std(df['Age']))
print("Grip Stren Mean:-", nmp.mean(df['Grip strength']), "  Standard Deviation:", nmp.std(df['Grip strength']))

print("\nCorrelation Analysis:")
hei_wei_corr = df['Height'].corr(df['Weight'])
print("Correlation of Height and Weight:-", hei_wei_corr)

age_grip_corr = df['Age'].corr(df['Grip strength'])
print("Correlation of Age and Grip Strength:-", age_grip_corr)

frailty_mean_weight = nmp.mean(df['Weight'][df['Frailty'] == 1])
non_frailty_mean_weight = nmp.mean(df['Weight'][df['Frailty'] == 0])
print("\nGroup-wise differnce:")
print("Mean Weight of Frailty:-", frailty_mean_weight)
print("Mean Weight of Non-Frailty:-", non_frailty_mean_weight)

t_test_weight, p_value_weight = stats.ttest_ind(df['Weight'][df['Frailty'] == 1], df['Weight'][df['Frailty'] == 0])
print("\nT-Test Results Weight:")
print("T-Statistic:-", t_test_weight)
print("P-Value:-", p_value_weight)
```
